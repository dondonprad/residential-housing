```python
merchant= {'grocery':['Indomaret','Alfamart','Idolmart','Superindo'],
           'fnb':['Mixue','Haus','Mie Gacoan','Sederhana','Fore Coffee','Kopi Kenangan'],
           'laundry':['Ceria','Si Bersih'],
           'retail':['Miniso','Nice So','Intermedia']
          }

group = []
name = []
merchant_id = []

for j in merchant.values():
    for k in j:
        name.append(k)
            
df_merchant = pd.DataFrame({'name_merchant':name})

merchant_id = ["{:02d}".format(i) for i in range(len(df_merchant['name_merchant'].values.tolist()))]
df_merchant['merchant_id'] = merchant_id

def corresponding_key(val, dictionary):
        for k, v in dictionary.items():
            if val in v:
                return k

group = []
for i in df_merchant['name_merchant'].values.tolist():
    group.append(corresponding_key(i,merchant))

df_merchant['group']=group

df_merchant
        
```

```python
# add review
import random
import pandas as pd
import datetime

def generate_random_date(start_date, end_date):
  """Generate a random date between two dates."""
  start_us = start_date.timestamp()
  end_us = end_date.timestamp()
  random_time = random.randint(start_us, end_us)
  return datetime.datetime.fromtimestamp(random_time)

def generate_synthetic_reviews():
  """Generates synthetic review data for various merchants."""

  categories = [i for i in df_merchant['merchant_group']]
  adjectives = ["baik", "buruk", "sangat baik", "sangat buruk"]
  nouns = ["pelayanan", "makanan dan minuman", "barang", "suasana"]
  review_templates = [
      " {noun} {adjective}.",
      "{noun}nya {adjective}, dimohon untuk {noun}nya ditingkatkan lagi.",
      "Saya mengalami pengalaman {adjective} secara keseluruhan.",
  
  ]

  data = []
  for category in categories:
    for _ in range(10):  # Generate 10 reviews per category
      merchant_name = f"{category} {random.randint(1, 1000)}"
      rating = random.randint(1, 5)

      # Generate review based on category
      if category == "fnb":
        nouns.extend(["penyajian", "suasana", "harga"])
      elif category == "grocery" and category == 'retail':
        nouns.extend(["produk", "harga", "kebersihan"])
      elif category == "laundry":
        nouns.extend(["kebersihan", "kualitas", "waktu pengerjaan"])

      review = random.choice(review_templates).format(
          adjective=random.choice(adjectives),
          noun=random.choice(nouns)
      )
      date = generate_random_date(datetime.datetime(2023, 1, 1), datetime.datetime(2024, 1, 1))

      # Remove category-specific nouns after generating review
      if category == "Restaurant":
        nouns.remove("penyajian")
        nouns.remove("suasana")
        nouns.remove("harga")
      elif category == "grocery" and category == 'retail':
        nouns.remove("produk")
        nouns.remove("harga")
        nouns.remove("kebersihan")
      elif category == "laundry":
        nouns.remove("kebersihan")
        nouns.remove("kualitas")
        nouns.remove("waktu pengerjaan")

      data.append([merchant_name, category, rating, review, date.strftime('%Y-%m-%d')])

  df = pd.DataFrame(data, columns=['merchant_name', 'category', 'rating', 'review', 'date'])
  return df

# Generate and save the dataset
df = generate_synthetic_reviews()
```

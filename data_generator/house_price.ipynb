```python
"""
Ancol A-H
Darmawangsa A-H
Kemayoran A-G
Menteng A-F
Senayan A-G
Thamrin A-G
"""
```

```python
clust = ['Kemayoran','Menteng', 'Ancol', 'Darmawangsa', 'Senayan', 'Thamrin' ]

# price reference
# Fungsi untuk menghasilkan harga acak dalam rentang tertentu
def generate_random_price(min_price, max_price):
  angka_awal = random.randint(min_price, max_price)
  angka_acak = str(angka_awal) + '000000'
  return int(angka_acak)

# Buat DataFrame
data = {'cluster': [], 'block': [], 'price': []}

# Isi DataFrame dengan data acak
for cluster in clust:
    for blok in ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']:
      
      if blok == 'A':
          harga = generate_random_price(350, 380)  # Sesuaikan rentang harga
      if blok == 'B':
          harga = generate_random_price(400, 420)
      if blok == 'c':
          harga = generate_random_price(430, 450)
      if blok == 'C':
          harga = generate_random_price(470, 500)
      if blok == 'D':
          harga = generate_random_price(460, 470)
      if blok == 'E':
          harga = generate_random_price(500, 520)
      if blok == 'F':
          harga = generate_random_price(530, 550)
      if blok == 'G':
          harga = generate_random_price(500, 510)
      if blok == 'H':
          harga = generate_random_price(480, 500)
          
      data['cluster'].append(cluster)
      data['block'].append(blok)
      data['price'].append(harga)

df_harga = pd.DataFrame(data)
df_harga
```

```python
df_harga = df_harga.drop([7,14,15,39,47])
df_harga.reset_index(inplace=True, drop=True)
df_harga
```

```python
clust_id = []

for i, j in zip(df_harga['cluster'], df_harga['block']):
    clust_join = i + '_' + j
    clust_id.append(clust_join)

clust_id
```

```python
cluster_code = ['01','02','03','04','05','06']

df_cluster_code = pd.DataFrame({'cluster_code':cluster_code,
                                'cluster_name':clust})

df_cluster_code
```

```python
df_harga['cluster_id'] = clust_id
df_harga
```

```python
clust_code = []

for i in df_harga['cluster']:
    ind = clust.index(i)
    cc = cluster_code[ind]
    clust_code.append(cc)

clust_code
```

```python
df_harga['cluster']=clust_code
df_harga
```

```python
df_harga = df_harga[['cluster_id','cluster','block','price']]
df_harga
```

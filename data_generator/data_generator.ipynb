# Create resident block data

```python
ancol = pd.read_csv('ancol.csv')
darma = pd.read_csv('darmawangsa.csv')
kemayoran = pd.read_csv('kemayoran.csv')
menteng = pd.read_csv('menteng.csv')
senayan = pd.read_csv('senayan.csv')
thamrin = pd.read_csv('thamrin.csv')
```

```python
df_resident = pd.concat([ancol, darma, kemayoran, menteng, senayan, thamrin], ignore_index=True)
df_resident
```

```python
rumah = ['Owner','Rent']
own_status_code = ['01','02']

cluster = ['Kemayoran','Menteng', 'Ancol', 'Darmawangsa', 'Senayan', 'Thamrin' ]
cluster_code = ['01','02','03','04','05','06']
```

```python
# add year of inhabited
tahun = [2022, 2023, 2024]
tmp_year = random.choices(tahun, weights=[316, 166, 307], k=789)

#add rent category
tmp_own = random.choices(rumah, weights=[671, 118], k=789)

# add year_of_inhabited & own_status
df_resident['year_of_inhabited'] = tmp_year
df_resident['own_status'] = tmp_own
```

```python
df_resident
```

```python
df_resident_list = df_resident.values.tolist()
```

```python
df_resident_list
```

```python
# add res_id
# res_id prefix = year_of_inhabited + cluster_code + number_block + own_status_code + random number (4 digit)
# total

yoi = []
clus_code = []
num_block = []
own_code = []
rand_num = []
```

```python
for i in df_resident_list:
  yoi.append(i[4])

for j in df_resident_list:
  ind = cluster.index(j[3])
  clus_code.append(cluster_code[ind])

for i in df_resident_list:
  num_block.append(i[2])

for k in df_resident_list:
  ind = rumah.index(k[5])
  own_code.append(own_status_code[ind])

for i in range(789):
  rand_num.append(str(uuid.uuid4().int)[:6])
```

```python
id_fix = []
id = np.c_[yoi, clus_code, num_block, own_code, rand_num]

for i in id:
  id_fix.append(''.join(i))
```

```python
id_fix
```

```python
df_resident['res_id'] = id_fix
df_resident
```

```python
df_resident.to_csv('resident_block.csv', index=False)
```

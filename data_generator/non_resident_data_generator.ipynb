```python
# add res_id
# res_id prefix = year_of_inhabited + cluster_code + number_block + own_status_code + random number (4 digit)
# because non-resident doesn't have house block, cluster data, using '0000' 4 digits 0 number
# then, the non-resident own_status using 00, totaly have 6 digits of 0 number. 

yoi = df_non_resident['year_of_inhabited'].values.tolist()
rand_num = []
id_fix = []

for i in range(500):
    rand_num.append(str(uuid.uuid4().int)[:6])
    
id_fix = []
id = np.c_[yoi, np.array(['00']*500), np.array(['00']*500), np.array(['00']*500), rand_num]

for i in id:
  id_fix.append(''.join(i))

id_fix
```

```python
df_non_resident['res_id'] = id_fix
df_non_resident
```

```python
# adding  work field and company 
work_field = ['perkebunan','perikanan dan peternakan','pertambangan, minyak, dan gas','pengolahan industri',
              'konstruksi','otomotif','logistik','teknologi dan informasi','aktifitas profesional, ilmiah, dan teknis',
              'keuangan dan asuransi','properti','perdagangan umum','Administasi dan manajemen','transportasi','pendidikan',
              'kesehatan', 'hukum', 'media', 'hiburan dan hospitality']

occup = ['karyawan swasta','PNS','pegawai BUMN','wirausaha','dokter','perawat','pengacara','wartawan','guru']


kerjaan = {
            'perkebunan':['karyawan swasta','pegawai BUMN', 'PNS', 'wirausaha'],
            'perikanan dan peternakan': ['wirausaha','PNS', 'karyawan swasta'],
            'pertambangan, minyak, dan gas' : ['PNS','karyawan swasta', 'pegawai BUMN'],
            'pengolahan industri': ['karyawan swasta', 'pegawai BUMN', 'wirausaha'],
            'konstruksi':['karyawan swasta', 'pegawai BUMN', 'PNS', 'wirausaha'],
            'otomotif':['karyawan swasta','wirausaha'],
            'logistik':['karyawan swasta', 'pegawai BUMN', 'wirausaha'],
            'teknologi dan informasi':['PNS','karyawan swasta', 'pegawai BUMN', 'wirausaha'],
            'aktifitas profesional, ilmiah, dan teknis': ['karyawan swasta', 'wirausaha'],
            'keuangan dan asuransi':['karyawan swasta', 'pegawai BUMN', 'wirausaha', 'PNS'],
            'properti':['karyawan swasta', 'wirausaha', 'PNS'],
            'perdagangan umum':['karyawan swasta', 'wirausaha', 'pegawai BUMN'],
            'Administasi dan manajemen':['karyawan swasta','pegawai BUMN', 'PNS'],
            'transportasi':['karyawan swasta', 'pegawai BUMN', 'wirausaha'],
            'pendidikan':['karyawan swasta', 'guru'],
            'kesehatan':['dokter', 'perawat', 'PNS', 'pegawai BUMN'],
            'hukum':['pengacara', 'karyawan swasta', 'pegawai BUMN', 'PNS'], 
            'media':['karyawan swasta', 'wartawan'],
            'hiburan dan hospitality':['karyawan swasta', 'wirausaha']
          }

rand_work_field = random.choices(work_field, weights=[5, 11, 15, 29, 13, 15, 12, 72, 10, 67, 49, 33, 70, 10, 23, 9, 40, 10, 7], k=500)

df_occu['work_field'] = rand_work_field

occupation = []
for i in df_occu['work_field']:
        occupation.append(random.choice(kerjaan[i]))
        
df_occu['occupation'] = occupation
df_occu
```

```python
title = ['staff','leader','spv','manager', 'senior manager', 'direktur']

title_rand = []
for i,j in zip(df_occu['occupation'], df_occu['age']):
    if i == 'wirausaha' and j<=52:
        title_rand.append('owner')
    elif i != 'wirausaha' and j <=34:
        title_rand.append(random.choice(title[:2]))
    elif i != 'wirausaha' and j >34 and j <=40:
        title_rand.append(random.choice(title[1:4]))
    elif i != 'wirausaha' and j>40:
        title_rand.append(random.choice(title[4:]))
                
title_rand
```

```python
df_occu['title'] = title_rand
df_occu
```

```python
#add salary 
# ['staff','leader','spv','manager', 'senior manager', 'direktur']

def salary_func(title_list):
    
    jobs = {
        # 'title': ( min salary, max salary)
        'staff': (4000000, 6599000),
        'leader': (6600000, 8499000),
        'spv': (8500000, 10499000),
        'manager': (10550000, 11999999),
        'senior manager': (12000000, 15499999),
        'direktur':(15500000, 50000000),
        'owner': (12000000, 15499000),
    }
    
    salary_list = []
    for i in title_list:
        min_salary = jobs[i][0]
        max_salary = jobs[i][1]
        salary = np.random.randint(min_salary, max_salary)
        salary_list.append(salary)
        
    return salary_list
```

```python
occu_salary = salary_func(title_rand)
occu_salary
```

```python
df_occu['income'] = occu_salary
df_occu
```

```python
child = np.random.randint(1,2)
child_num = []

for i in df_hh['marital_status']:
    if i == 'belum kawin':
        child_num.append(0)
    else:
        child_num.append(np.random.randint(1,2))

df_hh['child_num'] = child_num
df_hh
```

```python
has = ['yes', 'no']
has_motorcycle = []
has_car = []
for i in range (500):
    has_motorcycle.append(random.choice(has)) 
    has_car.append(random.choice(has))   
    
df_hh['has_motorcycle'] = has_motorcycle
df_hh['has_car'] = has_car
df_hh
```

```python
#adding use internet 
use_internet = random.choices(has,[445, 55], k=500)
df_hh['use_internet'] = use_internet
df_hh
```

```python
# adding household outcome / expense 
expense = []
for i in df_occu['income']:
    exp = int(i*(np.random.uniform(0.5,0.85)))
    expense.append(exp)

df_hh['expense'] = expense
df_hh
```

```python
# add work_by in df_occu 
#adding work by 
work_by = ['mobil', 'motor', 'transport umum']
work_by_rand = []

for i, j in zip(df_hh['has_motorcycle'], df_hh['has_car']):
    if i == 'yes' and j == 'yes':
        work_by_rand.append(random.choice(work_by))
    
    elif i == 'yes' and j == 'no':
        work_by_rand.append(random.choice(work_by[:2]))
    
    elif i == 'no' and j == 'yes':
        work_by_rand.append(random.choice(work_by[0:3:2]))
    else:
        work_by_rand.append('transportasi umum')
        
        
df_occu['work_by'] = work_by_rand
df_occu
```
